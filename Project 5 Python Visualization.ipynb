In this Jupyter Notebook, we willl attempt to recreate a GDP visualization in Pandas similar to that seen in Our World In Data (https://ourworldindata.org/grapher/gdp-per-capita-penn-world-table?tab=chart)

To do so we will take aggeragte compiling of each country and attempt to develop line plots for each country utilizing time series data. 

```python
#Importing Necessary Packages/Methods
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_gdp = pd.read_csv("gdp-per-capita-penn-world-table.csv")
```

```python
#Ensuring Dataframe has loaded properly
df_gdp.head(25)
```

Beginning to Develop Our Visualization

It is important to first notice that the visualizaton on Our World In Data has a filtering feature where you can select which countries to display. To best reflect the vizualation we will first filter/create a list of country to be included within the plot itself. The countries listed here were the ones first listed on the website when the page is loaded onto. 

```python
#Creation Of Countries List
countries = ["United States", "Germany", "United Kingdom", "France", "South Korea", "Japan", "China", "India"]
```

```python
#Creation of Data Frame
selected_countries = df_gdp[df_gdp["Entity"].isin(countries)]
US = selected_countries[selected_countries["Entity"]== "United States"]
plt.plot(US["Year"], US["GDP per capita (output, multiple price benchmarks)"])
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.title("GDP per Capita Over the Years: U.S")
plt.show()
```

Great! We can now see the GDP Per Capita of the United States over recent history. We could do a similar process to develop a line plot for each country and overlay this onto the same graph. To make matters simpler however, lets iterate over our created/selected dataframe and develop a line plot for each subsequent country within our specified countries list. ChatGPT reviewed as had slightly issue here at one point. 

```python
#For Loop Creation
for country in countries:
    country_gdp = selected_countries = df_gdp[df_gdp["Entity"] == country]
    plt.plot(country_gdp["Year"], country_gdp["GDP per capita (output, multiple price benchmarks)"], label=country)
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.title("GDP per Capita Over the Years")
plt.legend()
plt.show()
```

Wow! While all of the colors may not correspond/match from the Website Visualization to the One Here, things are quite similar! 

Also in just a couple changes, we can plot the same visualization in the Seaborn Package!

```python
for country in countries:
    country_gdp = selected_countries = df_gdp[df_gdp["Entity"] == country]
    sns.lineplot(x='Year', y='GDP per capita (output, multiple price benchmarks)', data=country_gdp, label=country)
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.title("GDP per Capita Over the Years")
plt.legend()
```
